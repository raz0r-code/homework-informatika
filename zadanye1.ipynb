# Решения задач
Уровни A, B, C

```python
# Уровень A
res = [n for n in range(1305, 14064) if (n % 2 == 0 or n % 3 == 0) and n % 7 != 0 and n % 11 != 0 and n % 17 != 0 and n % 23 != 0]
print(len(res), max(res))
```

```python
# Уровень B
res = [n for n in range(1529, 9483) if n % 4 == 1 and n % 5 == 3]
print(min(res), sum(res))
```

```python
# Уровень C
res = sorted([n for n in range(1005, 147871) if '1' not in str(n) and max(map(int, str(n))) - min(map(int, str(n))) < 4], reverse=True)
print(len(res), res[24])
```
